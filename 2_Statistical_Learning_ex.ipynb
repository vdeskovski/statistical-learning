```python
college_df['Top10perc'].hist()
```

```python
print("Mean =",college_df.Top10perc.mean())
print("Median =",college_df.Top10perc.median())
print("Min =",college_df.Top10perc.min())
print("Max =",college_df.Top10perc.max())
```

```python
top10perc_breakpoint = college_df.Top10perc.median()
```

```python
top10perc_max = college_df.Top10perc.max()
```

```python
# from 0 to median (exclusive) will be categorized as 'No', from median to the max value (inclusive) as 'Yes'
```

```python
college_df['Elite'] = pd.cut(college_df['Top10perc'], [0, top10perc_breakpoint, top10perc_max], labels=['No', 'Yes'])
```

```python
college_df[['Top10perc', 'Elite']]
```

```python
college_df.Elite.value_counts()
```

```python
college_df.Elite.value_counts().plot.bar()
```

```python
college_df.boxplot('Outstate', by='Elite')
plt.xlabel('Elite')
plt.ylabel('Outstate')
plt.show()
```

##### Higher Outstate tuition in the 'Elite' colleges as expected. A few outliers can be noticed in the NON 'Elite' colleges

### Percent of applicants accepted per college

```python
college_df['Acception Rate'] = college_df.Accept / college_df.Apps
```

```python
college_df['Acception Rate'] = college_df['Acception Rate'].round(3)
```

```python
college_df.head()
```

### Enrollment rate per college

```python
college_df['Enrollment Rate'] = college_df.Enroll / college_df.Accept
college_df['Enrollment Rate'] = college_df['Enrollment Rate'].round(3)
college_df.head()
```
